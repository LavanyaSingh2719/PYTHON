```python
#1. Load the 'tips' dataset from Seaborn and display the first 5 rows

df=sns.load_dataset('tips') #taking the tips dataset
df.head(5)
```

```python
#2. Create a scatter plot of total_bill vs tip using the 'tips' dataset.

sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('TotalBill vs Tips')
plt.show()
```

```python
#3. Plot a line graph for total_bill vs tip using Seaborn

sns.lineplot(x='total_bill',y='tip',data=df)
plt.title('TotalBill vs Tips')
plt.show()
```

```python
#4. Draw a bar plot to show the average tip for each day in the 'tips' dataset.

sns.barplot(x='day',y='tip',data=df)
plt.title('Average Tip by Day')
plt.show()
```

```python
#5. Use a count plot to display how many records are available for each day.

sns.countplot(x='day',data=df)
plt.title("Records")
plt.show()
```

```python
#6. Make a box plot to show total_bill distribution across different days

sns.boxplot(x='day',y='total_bill',data=df)
plt.title('Day-Wise TotalBill Distribution')
plt.show()
```

```python
#7. Create a violin plot for total_bill vs day.

sns.violinplot(x='day',y='total_bill',data=df)
plt.title('Violin Plot')
plt.show()
```

```python
#8. Plot a histogram of the 'total_bill' column with KDE using histplot.

sns.histplot(x='total_bill',data=df,bins=20,kde=True)
plt.title('Histogram')
plt.show()
```

```python
#20. Convert all categorical columns in the 'tips' dataset to numeric and create a full correlation heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('tips')
df.head()

df=pd.get_dummies(df,columns=['sex','smoker','time','day'])
corr=df.corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr,annot=True)
plt.title('Heatmap')
plt.show()
```
