```python
#Reading values of this dataset

import pandas as pd
df=pd.read_csv('Play_Store_Data.csv')
df
```

```python
#To find total missing values per column

#df.isnull().sum()
```

```python
#Clean the dataset: Handle missing values and duplicates.

df.info()#Gives an overview of the total missing values
print("")

df.isnull().sum() #To find total missing values in the dataset
print("")

df=df.dropna() #To remove missing values
df
```

```python
#Analyze ratings and their distribution.

df['Rating'].value_counts()
```

```python
#Identify most popular app category

df['Category'].value_counts().sort_values(ascending=False)#ascending is set to false so that category with highest value is shown at the top
```

```python
#Explore correlation between installs, reviews, and ratings.
```

```python
#Check app pricing trends (free vs paid).

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot: Count of Free vs Paid apps
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Type', palette='Set1')
plt.title('Number of Free vs Paid Apps')
plt.ylabel('Count')
plt.xlabel('App Type')
plt.show()
```

```python
#Visualize insights using bar plot

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar plot: Count of Free vs Paid apps
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Type', palette='Set2')
plt.title('Number of Free vs Paid Apps')
plt.xlabel('App Type')
plt.ylabel('Number of Apps')
plt.tight_layout()
plt.show()
```

```python
#Visualize insights using pie chart

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Pie chart: Free vs Paid app distribution
type_counts = df['Type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'orange'])
plt.title('Free vs Paid Apps Distribution')
plt.tight_layout()
plt.show()
```
